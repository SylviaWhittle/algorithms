```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
polygon_points = np.array([
    [1, 1],
    [2, 2],
    [2, 5],
    [3, 8],
    [6, 9],
    [9, 5],
    [8, 2],
    [6, 4],
    [5, 1],
    [1, 1]
])

```

```python
def point_in_polygon(point: np.ndarray, polygon: np.ndarray):
    n = polygon.shape[0]
    count = 0
    x = point[0]
    y = point[1]

    for index in range(polygon.shape[0]-1):
        x1, y1 = polygon[index, :]
        x2, y2 = polygon[index+1, :]

        if (y < max(y1, y2)) and (y >= min(y1, y2)):
            # if x is to the left of the intersection point.
            # x - x1 < (y - y1) / m
            # intersection's x-coord is x1 plus the difference of the point and 
            # p1's y coord, divided by the gradient.
            if x < (x2 - x1) * (y - y1) / (y2 - y1) + x1:
                count += 1
    if count % 2 == 0:
        return False
    else:
        return True


plt.plot(polygon_points[:, 0], polygon_points[:, 1])


points = []
for i in range(10):
    for j in range(10):
        points.append([i, j])

points = np.array(points)

for point in points:
    color = 'green' if point_in_polygon(point, polygon=polygon_points) else 'red'
    plt.plot(point[0], point[1], color=color, marker='.')
```
